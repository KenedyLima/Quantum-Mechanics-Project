```python
import numpy as np 
from qiskit import * #qiskit
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
from qiskit import Aer
from qiskit.extensions import Initialize
from qiskit import QuantumCircuit, QuantumRegister, transpile, execute
from qiskit.quantum_info import random_statevector
from qiskit.quantum_info import partial_trace, entropy
import qiskit.quantum_info as qi
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

# Gerando Ground State

```python
qr = QuantumRegister(2);
cr = ClassicalRegister(3);

circuit = QuantumCircuit(qr, cr);

# Definição dos parametros
k=1
h=1.5

#Preparação do ground state
alpha=-np.arcsin((1/np.sqrt(2))*(np.sqrt(1+h/np.sqrt(h**2+k**2))))

circuit.ry(2*alpha, qr[0])
circuit.cx(qr[0], qr[1])

#circuit.draw()
plot_bloch_multivector(circuit)
```

# Operações de Alice

```python
circuit.h(qr[0])
circuit.measure(qr[0], cr[0])
circuit.draw()
```

# Operações condicionais de Bob

```python
def sin(k,h):
    return (h*k)/np.sqrt((h**2+2*k**2)**2+(h*k)**2)

phi=0.5*np.arcsin(sin(k,h))

circuit.ry(2*phi,qr[1]).c_if(cr[0], 0)
circuit.ry(-2*phi,qr[1]).c_if(cr[0], 1)
circuit.h(qr[1])
circuit.measure(qr[1],cr[2])

circuit.draw()
```
