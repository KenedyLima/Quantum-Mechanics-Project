```python
## Imports
import qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.visualization import plot_bloch_multivector
from qiskit import IBMQ, Aer, transpile, assemble

qr = QuantumRegister(3, name="q")
cra = ClassicalRegister(1, name="cra")
crb = ClassicalRegister(1, name="crb")
circuit = QuantumCircuit(qr, cra, crb);

circuit.append(init, [0])
circuit.barrier()

circuit.h(1)
circuit.cnot(1, 2)
plot_bloch_multivector(circuit)
circuit.barrier()

circuit.cnot(0, 1)
circuit.h(0)

circuit.barrier()

circuit.measure(0, 0);
circuit.measure(1, 1)

circuit.barrier()

circuit.x(2).c_if(crb, 1)
circuit.z(2).c_if(cra, 1)
circuit.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
circuit.save_statevector()
out_vector = sim.run(circuit).result().get_statevector()
plot_bloch_multivector(out_vector)
```
